```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_csv = pd.read_csv('schedule.csv', delimiter='\t')
df_csv.head()
df_csv.info()
df_csv.describe()
df_csv.columns = ['Sr No', 'No of phases', 'Union Territories']
# print(df_csv.columns.tolist())
plt.figure(figsize=(8, 5))
sns.barplot(x='Union Territories', y='No of phases', data=df_csv, palette='viridis')
plt.title('The Schedule of GE to Lok Sabha- Bar Chart')
plt.xticks(rotation=80)
plt.xlabel('Union Territories')
plt.ylabel('No of phases')
plt.show()
```
